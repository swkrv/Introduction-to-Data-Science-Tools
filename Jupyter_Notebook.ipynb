## Arithmetic Expression Examples

In this section, we'll demonstrate some basic arithmetic operations commonly used in data science:

- **Addition:** Summing two numbers
- **Subtraction:** Finding the difference between two numbers
- **Multiplication:** Calculating the product of two numbers
- **Division:** Determining the quotient of two numbers

These fundamental operations are essential for data manipulation and analysis.


## Multiplying and Adding Numbers

### Define two numbers


```python
num1 = 10
num2 = 5
```


### Perform multiplication


```python
product = num1 * num2
```


### Perform addition


```python
sum_result = num1 + num2
```


### Print the results


```python

print("Product of", num1, "and", num2, "is:", product)
print("Sum of", num1, "and", num2, "is:", sum_result)
```

## Convert Minutes to Hours

### Define the number of minutes


```python
minutes = 150
```


### Convert minutes to hours


```python
hours = minutes / 60
```


### Print the result


```python
print(f"{minutes} minutes is equal to {hours:.2f} hours.")
```
